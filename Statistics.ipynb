## CATEGORICAL VS CATEGORICAL

### Wilderness_Area vs Cover_Type
H0: There is no association b/w Wilderness_Area vs Cover_Type             
H1: There is an association b/w Wilderness_Area vs Cover_Type

```python
tab=pd.crosstab(df['Wilderness_Area'],df['Cover_Type'])
tab
```

```python
a=stats.chi2_contingency(tab)
print('Chi2 is ',a[0])
print('P-value is ',a[1])
print('DOF is ',a[2])
print('Expected Values is ',a[3])
```

#### conclusion: Since the P-value(0)<alpha(0.05) rejecting H0 there is an association between the Wilderness_Area vs Cover_Type.

# CONTINOUS VS CATEGORICAL (One way Anova)

```python
## Assumptions : 1.Randomness 2.Normality 3.Variance equality
## 1: Assumed randomness
```

```python
##2.Normality
for i in ['Elevation','Aspect','Slope','Horizontal_Distance_To_Hydrology','Vertical_Distance_To_Hydrology','Horizontal_Distance_To_Roadways',
          'Hillshade_9am','Hillshade_Noon','Hillshade_3pm','Horizontal_Distance_To_Fire_Points','Distance_To_Hydrology', 'Hillshade']:
    a=stats.shapiro(df[i])
    if float(a[1])<0.05:
        print(i,' does not follow normality')
    else:
        print(i,' follow normality')
```

```python
#3.Variance equality
aa = df[df['Cover_Type']==1]
bb = df[df['Cover_Type']==2]
cc = df[df['Cover_Type']==3]
dd = df[df['Cover_Type']==4]
ee = df[df['Cover_Type']==5]
ff = df[df['Cover_Type']==6]
gg = df[df['Cover_Type']==7]
for i in ['Elevation','Aspect','Slope','Horizontal_Distance_To_Hydrology','Vertical_Distance_To_Hydrology','Horizontal_Distance_To_Roadways',
          'Hillshade_9am','Hillshade_Noon','Hillshade_3pm','Horizontal_Distance_To_Fire_Points','Distance_To_Hydrology', 'Hillshade']:
    abcd=stats.levene(aa[i],bb[i],cc[i],dd[i],ee[i],ff[i],gg[i])
    if float(abcd[1])<0.05:
        print('All input samples in from ', i,' does not have equal variances')
    else:
        print('All input samples in from ', i,' has equal variances')
```
